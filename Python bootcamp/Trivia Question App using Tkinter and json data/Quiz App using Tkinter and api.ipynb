```python
from tkinter import *

import requests

import pandas

import html
```

```python
class Quiz:
    def __init__(self):
        self.questions_data = self.generate_data()
        self.count_of_questions = 1
        
        
    def generate_data(self):
        parameters={
            "amount":10,
            "type":"boolean",
            "category":18
        }
        response = requests.get("https://opentdb.com/api.php",params=parameters)
        response.raise_for_status()
        self.data = response.json()
        
        #extracting the data
        self.q_data = self.data['results']
        self.ques_data={}
        i=1
        for item in self.q_data:
            self.ques_data[f'Q.{i}'] = {"question":item['question'],"answer":item['correct_answer']}
            i+=1
            
        return self.ques_data
    
    def still_has_questions(self):
        return self.count_of_questions < len(self.questions_data)
            

    def get_next_question(self):
        self.count_of_questions +=1
        return self.questions_data[f"Q.{self.count_of_questions-1}"]
        

    
```

```python

window =Tk()
window.config(height=800,width=500)
window.config(bg="#292E49",padx=50,pady=20)
window.title("Quizler")

quiz= Quiz()
score = 0
next_ques = {}

#functions

def update_Score():
    canvas.config(bg="#9ef0d2")
    global score_label,score
    score+=1
    score_label.config(text=f"Score : {score}")
    
def onclick_true():
    false_button.config(state="disabled")
    if next_ques['answer'] == 'True':
        update_Score()
    else:
        canvas.config(bg="#fe6583")
    window.after(1500,get_questions)
        
        
        
def onclick_false():
    true_button.config(state="disabled")
    if next_ques['answer'] == 'False':
        update_Score()
    else:
        canvas.config(bg="#fe6583")
        
    window.after(1500,get_questions)
        
    
def get_questions():
    canvas.config(bg="white")
    if quiz.still_has_questions():
        
        #enable the buttons after every new question
        true_button.config(state="normal")
        false_button.config(state="normal")
        
        global next_ques
        next_ques = quiz.get_next_question()
        question = html.unescape(next_ques['question'])
        canvas.itemconfig(question_display,text=question)
    else:
        
        canvas.itemconfig(question_display,text="You have reached the end.")
        true_button.config(state="disabled")
        false_button.config(state="disabled")
        




#canvas
canvas = Canvas(height=550,width=450)
canvas.grid(column=0,row=1,columnspan =2,pady=20)

#text 
question_display = canvas.create_text(250,250,text="Question goes here.",
                                      width=400,font=("Arial",15,"italic"),
                                      fill="#292E49")

#score
score_label = Label(text="Score: 0",bg="#292E49",fg="white",font=("Arial",15,"bold"))
score_label.grid(column=1 ,row=0)

#images
true_img = PhotoImage(file="right_mark.png")
false_img = PhotoImage(file="wrong_mark.png")

#buttons
true_button = Button(image=true_img,highlightthickness=0,command=onclick_true)
true_button.config(bg="#292E49")
true_button.grid(column=0,row=2)


false_button = Button(image=false_img,highlightthickness=0,command=onclick_false)
false_button.config(bg="#292E49")
false_button.grid(column=1,row=2)


    
get_questions()

window.mainloop()
```
