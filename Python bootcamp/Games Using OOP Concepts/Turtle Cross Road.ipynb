# Turtle crossing the road game using oop concepts and Turtle module.
    

```python
from turtle import Turtle,Screen
import time
import random


```

```python
class player(Turtle):
    def __init__(self):
        super().__init__()
        self.shape("turtle")
        self.color("black")
        self.penup()
        self.refresh()
        self.left(90)
        
    def move_up(self):
        self.forward(10)
        
    def refresh(self):
        self.goto(0,-230)
        
    
```

```python

class Scoreboard(Turtle):
    def __init__(self):
        super().__init__()
        self.level= 0
        self.penup()
        self.goto(-230,230)
        self.color("black")
        self.hideturtle()
        self.update_scoreboard()
        
    def update_scoreboard(self):
        self.clear()
        self.level +=1
        self.write(f"Level: {self.level}",align="left")
    
    
    def game_over(self):
        self.goto(0,0)
        self.write(f"Game Over")

        
        
        
        
        
        

        
```

```python
colors =["yellow","blue","red","green","violet","pink"]
STARTING_MOVE_DISTANCE = 5
SPEED_INCREMENT = 5
class carManager(Turtle):
    def __init__(self):
        self.all_cars=[]
        self.car_speed = STARTING_MOVE_DISTANCE
        
    def create_cars(self):
        random_creation = random.randint(1,6)
        if random_creation == 1:
            new_car = Turtle("square")
            new_car.shapesize(stretch_wid = 1,stretch_len=2)
            new_car.penup()
            new_car.color(random.choice(colors))
            random_y = random.randint(-200,200)
            new_car.goto(250,random_y)
            self.all_cars.append(new_car)
        
    def move(self):
        for car in self.all_cars:
            car.backward(self.car_speed)
            
    def level_up(self):
        self.car_speed += SPEED_INCREMENT
        
        
        
```

```python



screen = Screen()

screen.setup(height = 500, width = 500)
screen.title("Turtle crossing")
screen.tracer(0)
p1 = player()
car = carManager()
score = Scoreboard()

screen.listen()
screen.onkey(p1.move_up,"Up")




game_on =True

while game_on:
    time.sleep(0.1)
    screen.update()
    
    car.create_cars()
    car.move()
    
    #detect the collison with the cars
    for c in car.all_cars:
        if c.distance(p1) < 25:
            score.game_over()
            game_on = False
    
    # successful crossing
    if p1.ycor() >230:
        p1.refresh()
        car.level_up()
        score.update_scoreboard()
        


screen.exitonclick()
```
