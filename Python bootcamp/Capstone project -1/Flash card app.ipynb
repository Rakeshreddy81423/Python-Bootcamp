```python
from tkinter import *
import random
import pandas as pd
```

```python
BACKGROUND_COLOR = "#B1DDC6"
random_word ={}
words_dict={}


#import the french_eng data
try:
    translate_df = pd.read_csv("words_to_learn.csv")
except FileNotFoundError:
    translate_df = pd.read_csv("french_eng.csv")
    words_dict = translate_df.to_dict(orient="records")
else:
    words_dict = translate_df.to_dict(orient="records")



#flipping the card
def flip_card():
    canvas.itemconfig(title_text ,text="English")
    canvas.itemconfig(word_text,text=random_word['English'])
    canvas.itemconfig(card_background,image=card_front_img)
    canvas.config(bg="white",highlightthickness=1)
    

#button function
def next_word():
    global random_word, flip_timer
    window.after_cancel(flip_timer)
    random_word = random.choice(words_dict)
    canvas.itemconfig(title_text,text="French")
    canvas.itemconfig(word_text,text=random_word['French'])
    canvas.config(bg=BACKGROUND_COLOR, highlightthickness=0)
    flip_timer = window.after(3000,func=flip_card)
    
def is_known():
    words_dict.remove(random_word)
    data = pd.DataFrame(words_dict)
    data.to_csv("words_to_learn.csv",index=False)
    next_word()
    
    

window =Tk()
window.title("Flash card")
window.config(padx=50,pady=50,bg=BACKGROUND_COLOR)

flip_timer = window.after(3000,func=flip_card)





#canvas
canvas = Canvas(height=526,width=800)
card_front_img = PhotoImage(file="rectangle.png")
card_background = canvas.create_image(400,263,image=card_front_img)
canvas.config(bg=BACKGROUND_COLOR,highlightthickness=0,relief="raised")

title_text = canvas.create_text(400,200,text="French",font=("Arial",30,"italic"),fill="red")
word_text = canvas.create_text(400,265,text="Word",font=("Arial",30,"bold"),fill="green")
canvas.grid(column=0,row=0,columnspan=2)

#buttons

wrong = PhotoImage(file="wrong_mark.png")
wrong_button = Button(image=wrong,bg=BACKGROUND_COLOR,highlightthickness=0,command=next_word)
wrong_button.grid(row=1,column=0)

right = PhotoImage(file="right_mark.png")
right_button = Button(image=right,bg=BACKGROUND_COLOR,highlightthickness=0,command=is_known)
right_button.grid(row=1,column=1)


next_word()


window.mainloop()
```
