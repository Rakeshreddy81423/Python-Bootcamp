# Scraping the data from  https://stacker.com/stories/1587/100-best-movies-all-time  to get top 100 movies to watch

```python
from bs4 import BeautifulSoup
import requests

import html
```

```python
url="https://stacker.com/stories/1587/100-best-movies-all-time"

response = requests.get(url)

greatest_movies = response.text
soup = BeautifulSoup(greatest_movies,"html.parser")

all_movies = soup.find_all(name="h2",class_="ct-slideshow__slide__text-container__caption")
movie_names =[]

for movie in all_movies:
    movie_names.append(movie.getText().split("\n")[1].strip("#"))
    
top_100_movies = movie_names[1:][::-1]


for movie in top_100_movies:
    print(movie)



```
