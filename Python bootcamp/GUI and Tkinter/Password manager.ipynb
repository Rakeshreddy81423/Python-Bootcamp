# Password manager using tkinter 

In this project we create a app which helps us to generate random passwords to signup into any websites. And also save the details of website name, email and password into a file.

```python
from tkinter import *

from tkinter import messagebox

from random import randint,choice,shuffle

import json
```

```python
#window
window = Tk()
window.title("Password Manager")
window.config(padx =60,pady=60)


#save passwords to a file.
def onclick_add():
    website = website_entry.get().title()
    email = email_entry.get()
    password = password_entry.get()
    
    new_data = {
        website:{
        "email":email,
        "password":password
            }
    }
    
    if len(website) == 0 or len(password) == 0:
        messagebox.showinfo(title="Oops",message="Please fill all the fields.")
    else:
        try:
            with open("password_data.json","r") as data_file:
                #read old data
                data = json.load(data_file)
        except FileNotFoundError:
            #if file not found create a new json file and update the data
            with open("password_data.json","w") as data_file:
                json.dump(new_data,data_file,indent=4)
        else:
            #updating the new data
            data.update(new_data)
            
            with open("password_data.json","w") as data_file:
                json.dump(data,data_file,indent=4)
            
        finally:
            website_entry.delete(0,END)
            password_entry.delete(0,END)

#generate random password
def generate_random_password():
    letters = [chr(ascii_num) for ascii_num in range(65,91)] + [chr(ascii_num) for  ascii_num in range(97,123)]
    numbers = [chr(digit) for digit in range(48,58)]
    symbols = ['!','@','#','$','%','&','*','(',')','+','?']
    
    random_password = [choice(letters) for _ in range(randint(8,10))] +  [choice(numbers) for _ in range(randint(2,4))] + [choice(letters) for _ in range(randint(2,4))]
    shuffle(random_password)
    return "".join(random_password)
    
    

def onclick_generate_password():
    password_entry.delete(0,END)
    random_password = generate_random_password()
    password_entry.insert(0,random_password)
    
    
#search the email and passwords for a website name
def onclick_search():
    website_name = website_entry.get().title()
    try:
        with open("password_data.json","r") as data_file:
            #read old data
            data = json.load(data_file)
    except FileNotFoundError:
        messagebox.showinfo("Error","No previous saved passwords.")
    else:
        if website_name in data:
            messagebox.showinfo(title=f"{website_name}",message=f"Email: {data[website_name]['email']}\nPassword: {data[website_name]['password']}\n")
        else:
            messagebox.showinfo(title=f"Error", message=f"No details for {website_name} found")
            
        

    
   
        
        

#canvas for lock
canvas = Canvas(width=200,height=200)
lock_img = PhotoImage(file="lock.png")
canvas.create_image(100,100,image=lock_img)
canvas.grid(column=1,row=0)

#Labels
wesite_label = Label(text ="Website:")
wesite_label.grid(column=0,row=1)


email_label = Label(text="Email:")
email_label.grid(column=0,row=2)

password_label= Label(text="Password:")
password_label.grid(column=0,row=3)

#inputs using entry
website_entry = Entry(width=32)
website_entry.grid(column=1,row=1)
website_entry.focus()

email_entry = Entry(width=51)
email_entry.grid(column=1,row=2,columnspan=2)
email_entry.insert(0,"rakesh957306@gmail.com")

password_entry = Entry(width=32)
password_entry.grid(column=1,row=3)


#Buttons 

pswd_generate = Button(text="Generate Password",borderwidth=0.5,command = onclick_generate_password)
pswd_generate.grid(row=3,column=2)


add = Button(text="Add",width=44,borderwidth=0.5,command = onclick_add)
add.grid(column=1,row=4,columnspan=2)


search_button = Button(text="Search",borderwidth=0.5,command=onclick_search,width=14)
search_button.grid(row=1,column=2)






window.mainloop()
```
