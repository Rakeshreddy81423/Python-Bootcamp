```python
from tkinter import *
import math
```

```python
WORK_TIME = 25
SHORT_BREAK = 5
LONG_BREAK = 20
repeat = 0
timer = None


window= Tk()
window.title("Pomodoro")
window.config(padx=50,pady=50,bg="white")

#reset timer
def onclick_reset():
    window.after_cancel(timer)
    timer_label.config(text ="Timer")
    canvas.itemconfig(timer_text,text="00:00")
    global repeat
    repeat = 0

# timer mechanism
def onclick_start():
    global repeat
    repeat += 1
    
    if repeat % 2 == 0:
        check_marks.config(text = "âœ”"* int(repeat/2))
        
    if repeat % 8 == 0:
        timer_label.config(text="Break",fg = "red")
        count_down(LONG_BREAK * 60)
    elif repeat % 2 == 0:
        timer_label.config(text="Break",fg = "red")
        count_down(SHORT_BREAK * 60)
    else:
        timer_label.config(text="Work",fg = "green")
        count_down(WORK_TIME *  60)
        
        
        



def count_down(count):
    minutes = math.floor(count/60)
    seconds = count % 60
    if seconds < 10:
        seconds = f"0{seconds}"
    if minutes <10:
        minutes=f"0{minutes}"
        
    
    canvas.itemconfig(timer_text, text=f"{minutes}:{seconds}")
    if count > 0:
        global timer
        timer = window.after(1000,count_down,count-1)
    else:
        onclick_start()
        
    


timer_label= Label(text="Timer",bg= "white",font=("Arial",25,"bold"),fg="green")
timer_label.grid(column=1,row=0)

canvas = Canvas(width=200,height=200,bg="white",highlightthickness=0)
tomato_img = PhotoImage(file="tomato.png")
canvas.create_image(100, 80,image= tomato_img)
timer_text = canvas.create_text(95,107,text="00:00",fill="black",font=("Open Sans",20,"bold"))
canvas.grid(column=1,row=1)

start_button = Button(text="Start",highlightthickness=0,command=onclick_start)
start_button.grid(column=0,row=2)
reset_button = Button(text="Reset",highlightthickness=0,command= onclick_reset)
reset_button.grid(column=2,row=2)

check_marks = Label(text="",bg="white",fg="green")
check_marks.grid(column=1,row=3)

window.mainloop()
```
